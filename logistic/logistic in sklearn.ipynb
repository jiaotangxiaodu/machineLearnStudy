```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
X = X[y<2,:2]
y = y[y<2]
```

```python
plt.scatter(X[y==0,0],X[y==0,1],c='r')
plt.scatter(X[y==1,0],X[y==1,1],c='b')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
X_train,X_test,y_train,y_test = train_test_split(X,y)
log_reg.fit(X_train,y_train)
log_reg.score(X_test,y_test)
```

```python
probs = log_reg.predict_proba(X_test)[:,0]
```

```python
for i in range(0,len(X_test)):
    x_val = X_test[i,0]
    y_val = X_test[i,1]
    p = probs[i]
    plt.scatter(x_val,y_val,alpha=p,c='black')
plt.show()
```

```python
y_lines = (-log_reg.intercept_ - x_lines * log_reg.coef_[0][0])/(log_reg.coef_[0][1])
```

```python
for i in range(0,len(X_test)):
    x_val = X_test[i,0]
    y_val = X_test[i,1]
    p = probs[i]
    plt.scatter(x_val,y_val,alpha=p,c='black')
plt.plot(x_lines,y_lines)
plt.show()
```

```python
probs = log_reg.predict_proba(X_train)[:,0]
for i in range(0,len(X_train)):
    x_val = X_train[i,0]
    y_val = X_train[i,1]
    p = probs[i]
    plt.scatter(x_val,y_val,c=[[p,1-p,0]])
plt.plot(x_lines,y_lines)
plt.show()
```
