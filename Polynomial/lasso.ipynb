```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
x = np.random.uniform(-3,3,size=1000)
X = x.reshape(-1,1)
y = 0.5 * (x ** 2) + 1 * x + 2 + np.random.normal(0,1,1000)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
def LassoRegression(degree,alpha):
    return Pipeline([
        ("poly",PolynomialFeatures(degree=degree)),
        ("std_scalar",StandardScaler()),
        ("lasso_reg",Lasso(alpha=alpha))
    ])
```

```python
lasso1_reg1 = LassoRegression(degree=20,alpha=1e-2)
lasso1_reg1.fit(X_train,y_train)

lasso1_reg2 = LassoRegression(degree=20,alpha=1e-1)
lasso1_reg2.fit(X_train,y_train)

lasso1_reg3 = LassoRegression(degree=20,alpha=1e-0)
lasso1_reg3.fit(X_train,y_train)

lasso1_reg4 = LassoRegression(degree=20,alpha=1e1)
lasso1_reg4.fit(X_train,y_train)

lasso1_reg5 = LassoRegression(degree=20,alpha=1e2)
lasso1_reg5.fit(X_train,y_train)

lasso1_reg6 = LassoRegression(degree=20,alpha=1e3)
lasso1_reg6.fit(X_train,y_train)
```

```python
X_plt = np.linspace(-3,3,100).reshape(-1,1)
y_plt1 = lasso1_reg1.predict(X_plt)
y_plt2 = lasso1_reg2.predict(X_plt)
y_plt3 = lasso1_reg3.predict(X_plt)
y_plt4 = lasso1_reg4.predict(X_plt)
y_plt5 = lasso1_reg5.predict(X_plt)
y_plt6 = lasso1_reg6.predict(X_plt)
```

```python
plt.scatter(X_test,y_test,alpha=0.2)
plt.plot(X_plt,y_plt1,label='1')
plt.plot(X_plt,y_plt2,label='2')
plt.plot(X_plt,y_plt3,label='3')
plt.plot(X_plt,y_plt4,label='4')
plt.plot(X_plt,y_plt5,label='5')
plt.plot(X_plt,y_plt6,label='6')
plt.legend()
plt.show()
```
